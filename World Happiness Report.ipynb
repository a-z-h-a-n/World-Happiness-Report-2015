```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(' ', '_')
              .str.replace(r'[()]', '', regex=True)
)
```

```python
df.to_csv('world_happiness_2015_cleaned.csv', index=False)
```

```python
region_summary = df.groupby('region').mean(numeric_only=True).reset_index()
```

```python
region_summary.to_csv('world_happiness_2015_region_summary.csv', index=False)
```
